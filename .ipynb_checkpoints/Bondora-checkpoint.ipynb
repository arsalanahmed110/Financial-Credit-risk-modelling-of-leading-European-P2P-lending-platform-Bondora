```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Bondora_raw.csv', usecols=['ActiveLateCategory', 'ActiveLateLastPaymentCategory', 'ActiveScheduleFirstPaymentReached',
                                            'Age', 'Amount','AmountOfPreviousLoansBeforeLoan','AppliedAmount','City','ContractEndDate',
                                             'Country','FirstPaymentDate', 'HomeOwnershipType', 'IncomeTotal','InterestAndPenaltyBalance',
                                             'InterestAndPenaltyDebtServicingCost','InterestAndPenaltyPaymentsMade','InterestAndPenaltyWriteOffs',                                             
                                             'InterestRecovery','LastPaymentOn','LoanDate','LoanDuration', 'LoanId', 'LoanNumber', 'MonthlyPayment', 'MonthlyPaymentDay', 
                                             'UseOfLoan', 'Status'    
                                            ])
```

```python
df.columns
```

```python
percent_missing = df.isnull().sum() * 100 / len(df)
```

```python
percent_missing
```

```python
df = df.fillna(df.median())
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df1 = df.dropna(axis=1)
```

```python
df1.isna().sum()
```

```python
df1['Status']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encode = OneHotEncoder(sparse=False)
arr_encode = encode.fit_transform(df1[['Status']])
```

```python
df2 = pd.DataFrame(arr_encode, columns=['Repaid','Late','Remaining'])
```

```python
df2
```

```python
frames = [df1, df2]

final_df = pd.concat(frames)
```
